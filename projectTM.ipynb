```python
# CONVERT TO NEW DF PER WORD AND WITH POS TAG

def convert_to_new_df(df):
    data = []
    sentence_id = 1

    count = 0

    for index, row, in df.iterrows():
        text = row['text']
        if text.startswith('"') and text.endswith('"'):
            text = text[1:-2]
        labels = row['labels'].split()
        doc = nlp(text)

        words = text.split()
        pos = [token.tag_ for token in doc]

        if len(words) != len(labels):
            #print(f"Warning: Token count and label count do not match in row {index}")
            #print(words, labels)
            count += 1
            continue

        for i in range(len(words)):
            data_dict = {
                "Sentence #": f"{sentence_id}",
                "Word": words[i],
                "POS": pos[i],
                "Tag": labels[i]}
            data.append(data_dict)
        sentence_id += 1
    print('issues with', count, 'datapoints.', len(data), 'instances correctly converted')
    return pd.DataFrame(data)

ner_data = convert_to_new_df(ner_data)
ner_data.head(10)
```

```python
class SentenceGetter(object):
    
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, p, t) for w, p, t in zip(s["Word"].values.tolist(),
                                                           s["POS"].values.tolist(),
                                                           s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
    
    def get_next(self):
        try:
            s = self.grouped["Sentence: {}".format(self.n_sent)]
            self.n_sent += 1
            return s
        except:
            return None
```

```python
# input is a sentence as a structure show above 
#and and ith word from the sentence to return the features for that word

def word2features(sent, i):
    word = sent[i][0]
    postag = sent[i][1]
    
    # data structure consisting of a feature name and value for the token
    features = {
        'bias': 1.0,
        'word.lower()': word.lower(), # lower case variant of the token
        'word[-3:]': word[-3:], #suffix of 3 characters
        'word[-2:]': word[-2:], #suffix of 2 characters
        'word.isupper()': word.isupper(), # initial captial
        'word.istitle()': word.istitle(), # all words ini caps
        'word.isdigit()': word.isdigit(),
        'postag': postag,
        'postag[:2]': postag[:2], #first two characters of the PoS Tag
    }
    if i > 0:
        # adding features for the word based on the previous word
        word1 = sent[i-1][0] # previous word
        postag1 = sent[i-1][1]
        features.update({
            '-1:word.lower()': word1.lower(),
            '-1:word.istitle()': word1.istitle(),
            '-1:word.isupper()': word1.isupper(),
            '-1:postag': postag1,
            '-1:postag[:2]': postag1[:2],
        })
    else:
        features['BOS'] = True # Beginning of sentence as a feature

    if i < len(sent)-1:
        # adding features for the word based on the next word
        word1 = sent[i+1][0] # next word
        postag1 = sent[i+1][1]
        features.update({
            '+1:word.lower()': word1.lower(),
            '+1:word.istitle()': word1.istitle(),
            '+1:word.isupper()': word1.isupper(),
            '+1:postag': postag1,
            '+1:postag[:2]': postag1[:2],
        })
    else:
        features['EOS'] = True # end of sentence as a feature

    return features


def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]

def sent2labels(sent):
    return [label for token, postag, label in sent]

def sent2tokens(sent):
    return [token for token, postag, label in sent]
```

```python
sentences = getter.sentences
X = [sent2features(s) for s in sentences]
y = [sent2labels(s) for s in sentences]

total = len(X)
test_percent = 0.2
val_percent = 0.2 

test_size = int(total * test_percent) # 20% of the total data
trainval_size = total - test_size
val_size = int(trainval_size * val_percent) # so 20% of the training set
train_size = trainval_size - val_size

# Split while preserving order
X_train = X[:train_size]
y_train = y[:train_size]

X_val = X[train_size:train_size + val_size]
y_val = y[train_size:train_size + val_size]

X_test = X[-test_size:]
y_test = y[-test_size:]

# Display sizes
print(f"Train set size: {len(X_train)} sentences")
print(f"Validation set size: {len(X_val)} sentences")
print(f"Test set size: {len(X_test)} sentences")
```

```python
from itertools import product
```

```python
algorithms = ['lbfgs', 'l2sgd', 'ap', 'pa']
c1_values = [0.2, 0.1, 0.01] #L1 regularization
c2_values = [0.2, 0.1, 0.01] #L2
transitions = [True, False]
best_macro_f1_score = 0
best_f1 = 0
best_model_f1 = None
best_model_macro = None
best_macro_settings = {'c1': None, 'c2': None, 'algorithm': None, 'all_possible_transitions': None}
best_av_settings = {'c1': None, 'c2': None, 'algorithm': None, 'all_possible_transitions': None}


for alg, c1, c2, transition in product(algorithms, c1_values, c2_values, transitions):
    print(f'Training model with algorithm: {alg}, c1: {c1} and c2 {c2}, all_possible_transitions = {transition}')
    if alg == 'lbfgs':
        crf = CRF(algorithm=alg,
            c1=c1,
            c2=c2,
            max_iterations=100,
            all_possible_transitions=transition)
    elif alg == 'l2sgd':
        if c1 == 0.1 or c1 == 0.2:
            print('Skip because l2sgd does not support l1 regularization')
            continue
        else:
            c1 = None
            crf = CRF(algorithm=alg,
                c2=c2,
                max_iterations=100,
                all_possible_transitions=transition)
    else:
        if c1 == 0.1 and c2 == 0.1:
            c1 = None
            c2 = None
            crf = CRF(algorithm=alg,
                max_iterations=100,
                all_possible_transitions=transition)
        else:
            c1 = None
            c2 = None
            print('Skip because the algorithm does not support L1 and L2 regularization')
            continue

    crf.fit(X_train, y_train)
    y_pred = crf.predict(X_val)
    report = flat_classification_report(y_pred=y_pred, y_true=y_val, output_dict=True, zero_division=0)
    print(report)
    print()
    print('-----------------------------')
    f1_macro = report['macro avg']['f1-score']
    if f1_macro > best_macro_f1_score:
        best_macro_f1_score = f1_macro
        best_model_macro = crf
        best_macro_settings = {'c1': c1, 'c2': c2, 'algorithm': alg, 'all_possible_transitions': transition}
    f1 = report['weighted avg']['f1-score']
    if f1 > best_f1:
        best_f1 = f1
        best_model_f1 = crf
        best_av_settings = {'c1': c1, 'c2': c2, 'algorithm': alg, 'all_possible_transitions': transition}

```

```python
import csv
import pathlib
import sklearn
import numpy
import nltk
from nltk.corpus import stopwords
from collections import Counter
from sklearn.datasets import load_files
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
def parse_csv_like_json(filepath, id_key = 'sentence_id'):
    """
    The purpose of this function is to get the same format of data as the function json.load()
    but for csv/tsv files since this is the format of the training data/testing data
    -> Main problem was that the DictReader by itself is a flat dictionary so the data is not
    saved with their correct data types, so this would cause the analysis process to be harder
    for the VADER model
    """
    nested_data = {}
    with open(filepath, 'r', encoding='utf-8-sig') as f:
        #note: semicolon delimiter
        reader = csv.DictReader(f, delimiter=';') 
        for row in reader:
            #get ID and remove from row
            tweet_id = row.pop(id_key) 
            #convert fields to match JSON format
            tweet_info = {
                'sentiment_label': row['sentiment'],  #rename 'sentiment' to 'sentiment_label'
                'text_of_tweet': row['sentence'],     #rename 'sentence' to 'text_of_tweet'
                'topic': row['topic']                #keep 'topic' (add more fields if needed)
            }
            nested_data[tweet_id] = tweet_info
    return nested_data

#get train data
sentiment_train_data = parse_csv_like_json('train.csv')
```

```python
#load TSV test data
def parse_tsv_like_json(filepath):
    data = {}
    with open(filepath, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f, delimiter='\t')
        for row in reader:
            data[row['sentence_id']] = {
                'text_of_tweet': row['sentence'],
                'sentiment_label': row['sentiment'],
                'topic': row['topic']
            }
    return data

sentiment_test_data = parse_tsv_like_json('sentiment-topic-test.tsv')
```

```python
#prepare the data
#extract text and labels from train data
train_texts = [tweet['text_of_tweet'] for tweet in sentiment_train_data.values()]
train_labels = [tweet['sentiment_label'] for tweet in sentiment_train_data.values()]
#extract text and labels from test data
test_texts = [tweet['text_of_tweet'] for tweet in sentiment_test_data.values()]
test_labels = [tweet['sentiment_label'] for tweet in sentiment_test_data.values()]
```

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_validate
from itertools import product
import nltk
from nltk.corpus import stopwords

#initialize nltk resources
nltk.download('punkt')
nltk.download('stopwords')

#define parameter grid
min_df_values = [2, 5, 10]
ngram_ranges = [(1,1), (1,2), (2,2)]
use_idf_options = [True, False]
alpha_values = [10**x for x in [-3, -2, -1, 0]]

#best model tracking
best_macro_f1_score = 0
best_weighted_f1_score = 0
best_model_macro = None
best_model_weighted = None
best_macro_settings = {
    'min_df': None,
    'ngram_range': None,
    'use_idf': None,
    'alpha': None
}
best_weighted_settings = {
    'min_df': None,
    'ngram_range': None,
    'use_idf': None,
    'alpha': None
}

for min_df, ngram_range, use_idf, alpha in product(
    min_df_values, ngram_ranges, use_idf_options, alpha_values
):
    print(f'\nTraining with: min_df={min_df}, ngram_range={ngram_range}, '
          f'use_idf={use_idf}, alpha={alpha:.3f}')
    
    #create pipeline
    pipeline = Pipeline([
        ('vect', CountVectorizer(
            tokenizer=nltk.word_tokenize,
            stop_words=stopwords.words('english'),
            min_df=min_df,
            ngram_range=ngram_range
        )),
        ('tfidf', TfidfTransformer(use_idf=use_idf)),
        ('clf', MultinomialNB(alpha=alpha))
    ])
    
    #evaluate using cross-validation (5-fold)
    cv_results = cross_validate(
        pipeline,
        train_texts,
        train_labels,
        cv=5,
        scoring=['f1_macro', 'f1_weighted'],
        n_jobs=-1,
        verbose=1
    )
    
    #get average scores
    f1_macro = np.mean(cv_results['test_f1_macro'])
    f1_weighted = np.mean(cv_results['test_f1_weighted'])
    
    print(f"Macro F1: {f1_macro:.4f}, Weighted F1: {f1_weighted:.4f}")
    print('-----------------------------')
    
    #track best models
    if f1_macro > best_macro_f1_score:
        best_macro_f1_score = f1_macro
        best_model_macro = pipeline
        best_macro_settings = {
            'min_df': min_df,
            'ngram_range': ngram_range,
            'use_idf': use_idf,
            'alpha': alpha
        }
    
    if f1_weighted > best_weighted_f1_score:
        best_weighted_f1_score = f1_weighted
        best_model_weighted = pipeline
        best_weighted_settings = {
            'min_df': min_df,
            'ngram_range': ngram_range,
            'use_idf': use_idf,
            'alpha': alpha
        }

#final evaluation
print("\n=== Best Macro F1 Model ===")
print(f"Settings: {best_macro_settings}")
print(f"Macro F1: {best_macro_f1_score:.4f}")

print("\n=== Best Weighted F1 Model ===")
print(f"Settings: {best_weighted_settings}")
print(f"Weighted F1: {best_weighted_f1_score:.4f}")

#train final model on all training data and evaluate on test set
print("\n=== Training Final Model ===")
final_model = Pipeline([
    ('vect', CountVectorizer(
        tokenizer=nltk.word_tokenize,
        stop_words=stopwords.words('english'),
        min_df=best_weighted_settings['min_df'],
        ngram_range=best_weighted_settings['ngram_range']
    )),
    ('tfidf', TfidfTransformer(use_idf=best_weighted_settings['use_idf'])),
    ('clf', MultinomialNB(alpha=best_weighted_settings['alpha']))
])

final_model.fit(train_texts, train_labels)
y_pred_test = final_model.predict(test_texts)
print(classification_report(test_labels, y_pred_test, digits=3))
```

```python
#adapted conversion of the df from above to make data usable
def convert_tokenized_df(df):
    data = []
    count = 0

    # Group by sentence_id to process sentence-wise
    for sentence_id, group in df.groupby(0):
        if sentence_id == 'sentence_id':
            continue
        words = group[2].tolist()
        tags = group[3].tolist()
        for index, tag in enumerate(tags):
            if tag == 'B-ORG':
                tags[index] = 'B-org'
            elif tag == 'B-LOCATION':
                tags[index] = 'B-geo'
            elif tag == 'B-PERSON':
                tags[index] = 'B-per'
            elif tag == 'B-WORK_OF_ART':
                tags[index] = 'B-art'
            elif tag == 'I-LOCATION':
                tags[index] = 'I-geo'
            elif tag == 'I-ORG':
                tags[index] = 'I-org'
            elif tag == 'I-PERSON':
                tags[index] = 'I-per'
            elif tag == 'I-WORK_OF_ART':
                tags[index] = 'I-art'
        text = " ".join(words)

        doc = nlp(text)
        pos_tags = [token.tag_ for token in doc]

        # Check if spaCy's token count matches the original
        if len(doc) != len(words):
            count += 1
            continue

        for i in range(len(words)):
            data_dict = {
                "Sentence #": f"{sentence_id}",
                "Word": words[i],
                "POS": pos_tags[i],
                "Tag": tags[i]
            }
            data.append(data_dict)

    print(f"Issues with {count} sentence(s). {len(data)} tokens successfully processed.")
    return pd.DataFrame(data)

ner_test = convert_tokenized_df(ner_test)
ner_test.head(10)
```
