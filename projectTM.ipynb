```python
# CONVERT TO NEW DF PER WORD AND WITH POS TAG

def convert_to_new_df(df):
    data = []
    sentence_id = 1

    count = 0

    for index, row, in df.iterrows():
        text = row['text']
        if text.startswith('"') and text.endswith('"'):
            text = text[1:-2]
        labels = row['labels'].split()
        doc = nlp(text)

        words = text.split()
        pos = [token.tag_ for token in doc]

        if len(words) != len(labels):
            #print(f"Warning: Token count and label count do not match in row {index}")
            #print(words, labels)
            count += 1
            continue

        for i in range(len(words)):
            data_dict = {
                "Sentence #": f"{sentence_id}",
                "Word": words[i],
                "POS": pos[i],
                "Tag": labels[i]}
            data.append(data_dict)
        sentence_id += 1
    print('issues with', count, 'datapoints.', len(data), 'instances correctly converted')
    return pd.DataFrame(data)

ner_data = convert_to_new_df(ner_data)
ner_data.head(10)
```

```python
class SentenceGetter(object):
    
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, p, t) for w, p, t in zip(s["Word"].values.tolist(),
                                                           s["POS"].values.tolist(),
                                                           s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
    
    def get_next(self):
        try:
            s = self.grouped["Sentence: {}".format(self.n_sent)]
            self.n_sent += 1
            return s
        except:
            return None
```

```python
getter = SentenceGetter(ner_data)
```

```python
sent = getter.get_next()
```

```python
# input is a sentence as a structure show above 
#and and ith word from the sentence to return the features for that word

def word2features(sent, i):
    word = sent[i][0]
    postag = sent[i][1]
    
    # data structure consisting of a feature name and value for the token
    features = {
        'bias': 1.0,
        'word.lower()': word.lower(), # lower case variant of the token
        'word[-3:]': word[-3:], #suffix of 3 characters
        'word[-2:]': word[-2:], #suffix of 2 characters
        'word.isupper()': word.isupper(), # initial captial
        'word.istitle()': word.istitle(), # all words ini caps
        'word.isdigit()': word.isdigit(),
        'postag': postag,
        'postag[:2]': postag[:2], #first two characters of the PoS Tag
    }
    if i > 0:
        # adding features for the word based on the previous word
        word1 = sent[i-1][0] # previous word
        postag1 = sent[i-1][1]
        features.update({
            '-1:word.lower()': word1.lower(),
            '-1:word.istitle()': word1.istitle(),
            '-1:word.isupper()': word1.isupper(),
            '-1:postag': postag1,
            '-1:postag[:2]': postag1[:2],
        })
    else:
        features['BOS'] = True # Beginning of sentence as a feature

    if i < len(sent)-1:
        # adding features for the word based on the next word
        word1 = sent[i+1][0] # next word
        postag1 = sent[i+1][1]
        features.update({
            '+1:word.lower()': word1.lower(),
            '+1:word.istitle()': word1.istitle(),
            '+1:word.isupper()': word1.isupper(),
            '+1:postag': postag1,
            '+1:postag[:2]': postag1[:2],
        })
    else:
        features['EOS'] = True # end of sentence as a feature

    return features


def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]

def sent2labels(sent):
    return [label for token, postag, label in sent]

def sent2tokens(sent):
    return [token for token, postag, label in sent]
```

```python
sentences = getter.sentences
X = [sent2features(s) for s in sentences]
y = [sent2labels(s) for s in sentences]
```

```python
crf = CRF(algorithm='lbfgs',
          c1=0.1, #The coefficient for L1 regularization.
          c2=0.1, #The coefficient for L2 regularization.
          max_iterations=100,
          all_possible_transitions=False)
```

```python
pred = cross_val_predict(estimator=crf, X=X, y=y, cv=5)
report = flat_classification_report(y_pred=pred, y_true=y)
print(report)
```

```python
ner_test = pd.DataFrame(ner_test_data)
ner_test.head(10)
```

```python
#adapted conversion of the df from above to make data usable
def convert_tokenized_df(df):
    data = []
    count = 0

    # Group by sentence_id to process sentence-wise
    for sentence_id, group in df.groupby(0):
        if sentence_id == 'sentence_id':
            continue
        words = group[2].tolist()
        tags = group[3].tolist()
        for index, tag in enumerate(tags):
            if tag == 'B-ORG':
                tags[index] = 'B-org'
            elif tag == 'B-LOCATION':
                tags[index] = 'B-geo'
            elif tag == 'B-PERSON':
                tags[index] = 'B-per'
            elif tag == 'B-WORK_OF_ART':
                tags[index] = 'B-art'
            elif tag == 'I-LOCATION':
                tags[index] = 'I-geo'
            elif tag == 'I-ORG':
                tags[index] = 'I-org'
            elif tag == 'I-PERSON':
                tags[index] = 'I-per'
            elif tag == 'I-WORK_OF_ART':
                tags[index] = 'I-art'
        text = " ".join(words)

        doc = nlp(text)
        pos_tags = [token.tag_ for token in doc]

        # Check if spaCy's token count matches the original
        if len(doc) != len(words):
            count += 1
            continue

        for i in range(len(words)):
            data_dict = {
                "Sentence #": f"{sentence_id}",
                "Word": words[i],
                "POS": pos_tags[i],
                "Tag": tags[i]
            }
            data.append(data_dict)

    print(f"Issues with {count} sentence(s). {len(data)} tokens successfully processed.")
    return pd.DataFrame(data)

ner_test = convert_tokenized_df(ner_test)
ner_test.head(10)
```
